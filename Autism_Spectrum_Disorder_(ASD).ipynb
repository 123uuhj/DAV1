```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
asd = pd.read_csv("/content/Toddler Autism dataset July 2018.csv")
print("Dataset loaded...")
```

```python
d = pd.read_csv("/content/Toddler Autism dataset July 2018.csv")
print("Dataset loaded...")
```

```python
asd.head()
```

```python
sns.pairplot(d, kind="scatter")
plt.show()
```

```python
sns.set_style('darkgrid')
fig,ax=plt.subplots(nrows=5,ncols=2)
fig.set_size_inches(18.5,10.5)
sns.barplot(x='Sex',y="Age_Mons",data=asd,estimator=np.std,ax=ax[0,0]).set_title("bor plot")
sns.countplot(x='Sex',data=asd,ax=ax[0,1]).set_title("count plot")
sns.boxplot(x='Sex',y="Age_Mons",data=asd,hue="Qchat-10-Score",ax=ax[1,0]).set_title("box plot")
sns.violinplot(x='Sex',y="Age_Mons",data=asd,hue="Jaundice",split=True,ax=ax[1,1]).set_title("violin plot")
sns.stripplot(x='Sex',y="Age_Mons",data=asd,jitter=True,hue="Sex",dodge=True,ax=ax[2,0]).set_title("strip plot")
sns.swarmplot(x='Sex',y="Age_Mons",data=asd,ax=ax[2,1]).set_title("swarm plot")
sns.violinplot(x='Sex',y="Age_Mons",data=asd,ax=ax[3,0])
sns.swarmplot(x='Sex',y="Age_Mons",data=asd,color="black",ax=ax[3,0]).set_title("combined plot")
sns.boxplot(x='Sex',y="Age_Mons",data=asd,hue="Jaundice",ax=ax[3,1]).set_title("box plot")
sns.boxenplot(x="Sex",y="Age_Mons",color="b",scale="linear",data=asd,ax=ax[4,0])
sns.pointplot(x="Sex",y="Age_Mons",color="b",hue="Qchat-10-Score",data=asd,ax=ax[4,1])
```
